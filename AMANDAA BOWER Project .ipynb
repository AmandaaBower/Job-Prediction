```python
from getpass import getpass

COHERE_API_KEY = getpass()
```

```python
from langchain.llms import Cohere
from langchain import PromptTemplate, LLMChain
llm = Cohere(cohere_api_key=COHERE_API_KEY)
```

```python
from langchain.prompts import PromptTemplate
```

```python
from langchain.chains import LLMChain
```

```python
template = """
Interprete the text and evaluate the text.
course:write a description about mentioned course ?
Description:What is the description of the subject? Use few lines of words.

Format the output with the following keys:
course
description

text: {input}
"""

prompt_template = PromptTemplate.from_template(template=template)
chain = LLMChain(llm=llm, prompt=prompt_template)
chain.predict(input="Write a description about data science")
```

```python
response_template = """
You are a helpful bot that generate a 'road map for the course based on the skills' reponse text.
Road map should be based on the course core skills like coding(python,R language),statistic&probability,visualization etc...
You will get a course and description as into and evaluate.

text: {input}
"""
review_template = PromptTemplate(input_variables=["input"], template=response_template)
review_chain = LLMChain(llm=llm, prompt=review_template)
review_chain.predict(input="Generate a road map for data science and present it in step-by-step manner.")
```

```python
from langchain.chains import SimpleSequentialChain

overall_chain = SimpleSequentialChain(chains=[chain, review_chain], verbose=True)

overall_chain.run(input="Write a description about data science and generate a road map for data science and present it in step-by-step manner")
```
